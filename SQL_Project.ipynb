TRIGGER

```python
# Create trigger called adv_drug
# After inserting a new row into prescriptions
# For each new row...
# Start trigger body
# define earlier_drug 
# Place drug_name into earlier drug 
# From the prescriptions table
# Where the patient_id matches in both cases
# And drug_name is created as follows
# Choose drug_name_1
# From the adverse_interactions table
# Where the second drug name and new drug name are the same
# If there is an earlier drug and the above conditions are met
# Insert new tuple into alerts table 


# query = "CREATE TRIGGER adv_drug \
# AFTER INSERT ON prescriptions \
# FOR EACH ROW \
# BEGIN \
#     DECLARE earlier_drug VARCHAR(128); \
# 	SELECT drug_name INTO earlier_drug \
#     FROM prescriptions \
#     WHERE patient_id = NEW.patient_id \
#       AND (drug_name IN ( \
#           SELECT drug_name_1 \
#           FROM adverse_reactions \
#           WHERE drug_name_2 = NEW.drug_name \
#       ) \
#       OR drug_name IN ( \
#           SELECT drug_name_2 \
#           FROM adverse_reactions \
#           WHERE drug_name_1 = NEW.drug_name \
#       )); \
# 	IF earlier_drug IS NOT NULL THEN \
# 		INSERT INTO alerts (patient_id, physician_id, alert_date, drug1, drug2) \
#         VALUES (NEW.patient_id, NEW.physician_id, NEW.date, earlier_drug, NEW.drug_name); \
# 	END IF; \
# END;"


# CORRECT TRIGGER BELOW

query = "CREATE TRIGGER adv_drug \
AFTER INSERT ON prescriptions \
FOR EACH ROW \
BEGIN \
    INSERT IGNORE INTO alerts (patient_id, physician_id, alert_date, drug1, drug2) \
    SELECT Distinct NEW.patient_id, NEW.physician_id, NEW.date, drug_name, NEW.drug_name \
    FROM prescriptions \
    WHERE patient_id = NEW.patient_id \
      AND (drug_name IN ( \
          SELECT drug_name_1 \
          FROM adverse_reactions \
          WHERE drug_name_2 = NEW.drug_name \
      ) \
      OR drug_name IN ( \
          SELECT drug_name_2 \
          FROM adverse_reactions \
          WHERE drug_name_1 = NEW.drug_name \
      )) ; \
END;"



mycursor.execute(query)
mydb.commit()
```
